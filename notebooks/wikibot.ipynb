```python
import pywikibot
import mwparserfromhell
```

```python
site = pywikibot.Site('en', 'wikipedia')  # The site we want to run our bot on
page = pywikibot.Page(site, 'American_logistics_in_the_Normandy_campaign')
text = page.get()
parsed = mwparserfromhell.parse(text)

parsed
```

```python
links = []

for node in parsed.filter_wikilinks():
    links.append(node)

[str(link.title) for link in links]
```

```python
links[1]
```

```python
def is_valid_link(link):
    # Filter out images, categories, and non-internal links
    return not str(link.title).startswith(("File:", "Category:", "Talk:", "User:", "Template:", "Help:", "Portal:", "Draft:", "TimedText:", "MediaWiki:", "Module:", "Special:"))

links = [link for link in links if is_valid_link(link)]

[str(link.title) for link in links]
```
